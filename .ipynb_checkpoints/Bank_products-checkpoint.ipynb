### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features, prepare the features and target column for modeling with appropriate encoding and transformations.

### Identify Bank Related Features

````python
# Bank-specific features (excluding economic indicators)
bank_features = [
    'age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
    'contact', 'month', 'day_of_week', 'duration', 'campaign',
    'pdays', 'previous', 'poutcome'
]

# Prepare X and y
X_bank = df_clean[bank_features].copy()
y = df_clean['y'].copy()

print(f"Selected feature columns:\n{bank_features}")
print("\nSample data:")
display(X_bank.head())
````

### Encode Binary Features Using Label Encoding

````python
# Binary categorical columns
binary_cols = ['default', 'housing', 'loan']

# Initialize dictionary to store encoders (if needed for inverse transform later)
label_encoders = {}

# Apply label encoding to binary features
for col in binary_cols:
    le = LabelEncoder()
    X_bank[col] = le.fit_transform(X_bank[col])
    label_encoders[col] = le
````

### One-Hot Encode Remaining Categorical Columns

````python
# Categorical columns for one-hot encoding
onehot_cols = ['job', 'marital', 'education', 'contact', 'month', 'day_of_week', 'poutcome']

# Define ColumnTransformer
ct = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(drop='first', sparse_output=False), onehot_cols)],
    remainder='passthrough'  # Keep non-encoded columns as-is
)

# Fit-transform the data
X_encoded = ct.fit_transform(X_bank)
````

### Encode the Target Variable

````python
# Encode target (yes/no) to binary
target_le = LabelEncoder()
y_encoded = target_le.fit_transform(y)

# Show positive class percentage
positive_class_pct = round((y_encoded.sum() / len(y_encoded)) * 100, 2)
print(f"\nPositive class percentage: {positive_class_pct}%")
````

````python
# Get feature names from one-hot encoding
ohe = ct.named_transformers_['onehot']
onehot_feature_names = ohe.get_feature_names_out(onehot_cols)

# Get passthrough (non-onehot) feature names
passthrough_cols = [col for col in X_bank.columns if col not in onehot_cols]

# Final column names in correct order
all_feature_names = list(onehot_feature_names) + passthrough_cols

# Convert result to DataFrame
X_encoded_df = pd.DataFrame(X_encoded, columns=all_feature_names)

# Preview
print("\nEncoded feature set (first 5 rows):")
display(X_encoded_df.head())

# Check class distribution
print("\nTarget class distribution:")
print(pd.Series(y_encoded).value_counts(normalize=True))
````

A notable observation from the displayed rows is that the pdays column consistently shows 999.0, which often signifies that the client was not previously contacted or a large number of days passed since the last contact. Crucially, the "Target class distribution" section reveals a significant class imbalance:

Class 0: Represents approximately 88.73% of the data.
Class 1: Represents approximately 11.27% of the data.

This imbalance is a critical factor for any machine learning task on this dataset, as it suggests that models might be biased towards predicting the majority class (Class 0) if not properly addressed.

### Problem 6: Train/Test Split

You now have a clean, one-hot and label-encoded feature set using only bank-related data. This is ready for initial modeling and analysis.

````python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded_df,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

# Print shapes and class distribution
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)

print("\nTrain class distribution (%):")
print(pd.Series(y_train).value_counts(normalize=True) * 100)

print("\nTest class distribution (%):")
print(pd.Series(y_test).value_counts(normalize=True) * 100)

# Now scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

To split a clean, one-hot and label-encoded feature set (bank-related data) into training and testing subsets for initial modeling.
Method: Train-test split is performed using train_test_split with test_size=0.2 and random_state=42.
stratify=y_encoded is used to ensure that the class distribution of the target variable (y_encoded) is preserved in both the training and testing sets.

Class Distribution: Both training and test sets show a very similar class distribution (approximately 88.73% for class 0 and 11.27% for class 1), confirming the success of stratification

````python
# Define models with default settings
models = {
    'Logistic Regression': LogisticRegression(max_iter=2000, random_state=42),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'SVM': SVC(random_state=42, probability=False)  # Set probability=True if needed later
}

results = []

for name, model in models.items():
    # 1. Apply SMOTE to training data
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

    # 2. Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_smote)
    X_test_scaled = scaler.transform(X_test)

    # 3. Time model fitting
    start_time = time.time()
    model.fit(X_train_scaled, y_train_smote)
    train_time = time.time() - start_time

    # 4. Calculate accuracies
    train_acc = model.score(X_train_scaled, y_train_smote)
    test_acc = model.score(X_test_scaled, y_test)

    # 5. Store results
    results.append({
        'Model': name,
        'Train Time (s)': round(train_time, 3),
        'Train Accuracy': round(train_acc, 3),
        'Test Accuracy': round(test_acc, 3)
    })

# Create DataFrame of results
comparison_df = pd.DataFrame(results)
print(comparison_df)
````

Training Time:
KNN trained exceptionally fast (0.006 seconds), while SVM took significantly longer (39.884 seconds), making it the slowest among the models. Logistic Regression and Decision Tree had moderate training times, with Decision Tree being slower than Logistic Regression but faster than SVM.

Training Accuracy:
The Decision Tree achieved perfect training accuracy (1.000), which strongly suggests overfittingâ€”it memorizes the training data but may fail to generalize. KNN and SVM both showed high training accuracy (0.951 and 0.950 respectively), indicating good fit to the training data without reaching perfection.

Test Accuracy:
SVM achieved the highest test accuracy (0.903), demonstrating strong generalization despite its longer training time. Logistic Regression followed closely (0.856), providing a good balance between accuracy and efficiency. Decision Tree, despite perfect training accuracy, had a comparatively lower test accuracy (0.873), reinforcing the overfitting concern. KNN had the lowest test accuracy (0.840), trading some accuracy for very fast training.

SVM offers the best test performance but at a high computational cost.
Logistic Regression is the most balanced model, combining reasonable training time with solid test accuracy.
Decision Tree shows signs of overfitting, with perfect training but lower test accuracy.
KNN trains extremely fast but sacrifices some accuracy on unseen data.

````python
# Feature engineering
# Interaction feature: age * job code
df_clean['age_job_interaction'] = df_clean['age'] * pd.Categorical(df_clean['job']).codes

# Binning age into groups
bins = [0, 25, 45, 65, 100]
labels = ['Young', 'Adult', 'Senior', 'Elder']
df_clean['age_group'] = pd.cut(df_clean['age'], bins=bins, labels=labels)

# Map month to quarter
month_to_quarter = {
    'jan': 1, 'feb': 1, 'mar': 1,
    'apr': 2, 'may': 2, 'jun': 2,
    'jul': 3, 'aug': 3, 'sep': 3,
    'oct': 4, 'nov': 4, 'dec': 4
}
df_clean['quarter'] = df_clean['month'].str.lower().map(month_to_quarter)

# Map day_of_week to weekend vs weekday binary flag
weekend = ['sat', 'sun']
df_clean['is_weekend'] = df_clean['day_of_week'].str.lower().isin(weekend).astype(int)

# Show the first few rows with new features
print(df_clean.head())
````

* An age_job_interaction feature is created by multiplying the age column with the numerical codes derived from the categorical job column.
  
* The numerical age column is binned into four categorical groups: 'Young' (0-24), 'Adult' (25-44), 'Senior' (45-64), and 'Elder' (65-100), stored in a new age_group column.

* The month column is mapped to its corresponding quarter (1-4) using a predefined dictionary.
* A binary is_weekend flag is generated, indicating whether a day_of_week is a weekend (1 for Saturday/Sunday) or a weekday (0).

The print(df_clean.head()) command shows the first five rows of the DataFrame, confirming the successful addition of these new features (age_job_interaction, age_group, quarter, is_weekend) alongside the original and potentially other engineered columns, resulting in a DataFrame with 32 columns.

### Integrate SMOTE with Pipeline
Use imblearnâ€™s pipeline to apply SMOTE only on training folds inside CV:

````python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier

pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', DecisionTreeClassifier(random_state=42))
])
````

### Hyperparameter Tuning with GridSearchCV

````python
df_analysis = df_clean.copy()
df_analysis.rename(columns={'y': 'target'}, inplace=True)
````

````python
param_grid = {
    'classifier__max_depth': [3, 5, 10, 15, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(df_analysis.drop('target', axis=1), df_analysis['target'])

print("Best parameters:", grid_search.best_params_)
print("Best CV accuracy:", grid_search.best_score_)

````

The model achieved its highest cross-validation accuracy of 73.4% using the above hyperparameters. Allowing unlimited tree depth, while requiring each split and leaf to have at least 2 samples, helped the model balance complexity and generalization. These parameters can be used as a baseline for further model improvement or evaluation.

````python
best_model = grid_search.best_estimator_.named_steps['classifier']

# Get feature names from preprocessor
ohe = grid_search.best_estimator_.named_steps['preprocessor'].named_transformers_['cat']
ohe_features = ohe.get_feature_names_out(categorical_cols)
all_features = numeric_cols + list(ohe_features)

importances = best_model.feature_importances_
feat_imp_df = pd.DataFrame({
    'Feature': all_features,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feat_imp_df.head(10))
````

"Duration" is by far the most important feature, followed by "pdays" and "age."
